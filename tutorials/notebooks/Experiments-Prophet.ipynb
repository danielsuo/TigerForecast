```python
%matplotlib inline
import tigerforecast as tf 
import matplotlib.pyplot as plt
```

```python
ar = tf.problems.ARMA()
m1, m2, m3 = tf.methods.PredictZero(), tf.methods.LastValue(), tf.methods.Prophet()
x = ar.initialize()
m1.initialize(), m2.initialize(), m3.initialize()

T, loss1, loss2, loss3 = 50, [], [], []
for _ in range(T):
    x1, x2, x3 = m1.predict(x), m2.predict(x), m3.predict(x)
    x = ar.step()
    loss1.append((x - x1)**2)
    loss2.append((x - x2)**2)
    loss3.append((x - x3)**2)
```

```python
plt.plot(loss1, c='red')
plt.plot(loss2, c='blue')
plt.plot(loss3, c='green')
```
