```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import medfilt
```

```python
ys = np.load('ys.npy')
offline = np.load('residual.npy')
online = np.load('online_residual.npy')
```

```python
print(online.mean())
```

```python
plt.rc('figure', figsize=(12,12))

plt.subplot(311)
plt.plot(ys, 'k')
plt.grid(zorder=-1)

plt.ylabel('Mean discharge')

plt.subplot(312)
plt.ylabel('Prediction residuals')
plt.semilogy(medfilt(offline), 'b', label='offline')
plt.semilogy(medfilt(online), 'c', label='online')
plt.grid(zorder=-1)

plt.legend()

plt.subplot(313)

plt.ylabel('Prediction advantage (online - offline)')
plt.plot(online - offline, 'r')
plt.axhline(y=0, c='k')
plt.grid(zorder=-1)
plt.ylim([-0.1, 0.1])

plt.xlabel('Time')

plt.tight_layout()
```
